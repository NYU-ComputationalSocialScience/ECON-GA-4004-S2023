## Basic Setting
We study the dynamic labor economic model proposed by Burdett and Mortensen(1977).
This is very different from Mccall model we studied before.Bascially, we include search, job seperation and recall. 
In the set up of Burdett and Mortensen's model, they have the following assumptions:

-An individual can occupy the following labor market states:

-1 $u$.Unemployment - that is, without a job, actively seeking work, and not subject to recall
from a formerly held job

-2 $e(w)$ Employment at wage $w$. In this case the individual is attached to some firm that pays
her a wage of $w$.

-3 $d(w)$ Laid-off from a job that paid a wage of $w$. In this case the individual was “temporarily
dismissed” from a job that had paid her a wage of $w$.

In their paper, they also include out of the labor force (OLF). In this case the individual is out of the labor market and do not actively search for work. However there is no need to include OLF as the individual is no longer in the labor market, and it could be regarded as a special case of unemployment when the searching intensity equals to zero.

 Also, we have a few parameters and assumptions needed to be set up, just like MCcall model:
 
 -The wage offers follow an i.i.d draw from the wage distribution $F(w_j)$ where $0<w_1<\ldots<w_n$. In every period of searching, the individual will face one wage offer drawn from the distribution when she finds a firm. The distribution is known by the population so they can make rational expecations.
    
 -Every individual will have one unit of time per period $t$, while there could be infinite continuous time. In our model we choose to discretinize both the time and the wage distribution.
    
 -Individual of all states could search, and those who decide to search must determine a searching intensity $s, s\in [0,1]$ with a successful rate $\lambda$. So an individual could meet a new firm with probability $\lambda s$.
    
 -All employed individuals are separated from their current employers (at a risk of losing job) at a rate $\eta$. This rate applies to both emplyed and laid-off workers.
    
 -All individuals actively employed at any firm are subject to a risk of layoff. Layoff occurs at rate $\zeta$.
    
 -All laid off individuals are subject to a recall to their job with a rate $\upsilon$.
    
 -Individuals have a contemporaneous utility function given by $$U(c,l)$$ where $c$ is the instant consumption (represented by income due to no capital market) and $l$ is the leisure time.
    
 -All individuals have a nonlabor income flow of $y > 0$. This $y$ applies to people at all states.
    
 -Within the one unit of time, individual could choose to search or relax, while the working time is the rest of them: $$h=1-s-l$$and the income for employees at a given wage $w$ is a function of working time: $c=w(h)+y$.
    
 -The intertemporal discount rate is given by $\beta \in (0,1)$

## Structure of the Optimization Problem
The crucial difference at BM model is that, all individuals arrive the labor market in the three states, and they do not observe the wage offer immediately (while in MCcall model, we assume that all individuals will observe their wage $w$ without doing anything), and they need to spend some time searching for the job--No matter unemployed, employed or laid-off. That is to say, if the unemployed individual does not search, she will not observe any new wage offer and stay unemployed until next period, while in MCcall model the wage offers are 'presented' to her with no cost.

The time needed to search for new offers, or the time length of a period could be very short, let $\epsilon$ be the length of each time period $t$.

Let us start looking at problem faced by an individual who is currently unemployed. The Bellman equation is given by:
$$V_u(y)=\max_{s}(1+\beta\epsilon)^{-1}\{U(y,1-s)\epsilon+\lambda s \epsilon \sum_{i=1}^{n}q_i \max [V_u(y),V_e(w;y)]+(1-\lambda s\epsilon)V_u(y)+o(\epsilon)\}$$
The Bellman equation indicates that, for different exogenous income $y$, there should be different level for the value of employement and unempolyment. For one fix value of $y$, we can ignore it, but now let us keep it first.

The interpretation of the Bellman equation is the following: $\epsilon$ should be considered as a very short time length for the time period, over which there could be approximately nothing happening. The unemployed individual will get her income/consumption $y$ and comnsume it all, and in addition, she will determine the intensity of searching for a new job. If she decides to use $s$ time for searching, she will get $1-s$ leisure time and the total utility will become $U(y,1-s)$, and this instantaneous utility level will be attained for time period $\epsilon$, so in the short time period $\epsilon$, the total utility will be $U(y,1-s)\epsilon$. Since this happens after $\epsilon$, we need to discount back to the beginning, which is now, so we have the discount factor for this time period: $$\frac{1}{1+\beta\epsilon}$$
So the current value of total utility discounted back from $\epsilon$ after now is $(1+\beta\epsilon)^{-1}U(y,1-s)\epsilon$\
For the searching activity, since in the BM model, searching event is described as a Poisson process, and by the Markov property of Poisson process,the successful rate for searching is correlated with the time spent on searching. As in the basic setup we described, in one unit of time with intensity (or proportion spent on searching) $s$, the successful rate is $\lambda s$, so in $\epsilon$ time period the successful rate would be $\lambda s \epsilon$.

Because this Poisson process could be described as a two discrete states process, the outcome of searching is relative simple: meet a new firm or not. If the unemployed individual meets a new offer drawn from the wage distribution $F(w)$, she will immediately compare the value of employement and the value of unemployement, $V_e(w;y)$ and $V_u(y)$. If the wage is high enough, she will accept the offer or vice versa. Actually, since individuals has complete information about the wage distribution, they can form a rational expectation for the choice of accepting the firms or not: for each possible wage $w_j$ with probability $q_j$, the unemployed individual will make such comparison and choose the maximum between $V_e(w;y)$ and $V_u(y)$. Thus, like the one in MCcall model, the expected value for a successful search before observing the wage is $\sum_{i=1}^{n}q_i \max [V_u(y),V_e(w;y)]$.
Apparently, if the search fails, the individual will stay unemployed with value of $V_u(y)$

Lastly, we have $o(\epsilon)$. This term is to describe some very rare cases that may happend in a short period of time, such as receiving multiple offers at the same time, and it has the following propery: $$\lim_{\epsilon \to 0}\frac{o(\epsilon)}{\epsilon}=0$$
Generally speaking, we do not think the event in $o(\epsilon)$ to be highly possible.

## The Bellman Equation System for the Model 

With assumptions and setup above, we are very close the the ultimate form of BM model, but before writing down the Bellman equations for employed and laid-off, we should first rearrange the equation for unemployment.

Mulitplying through by $(1 + \beta \epsilon)$, we get
$$(1 + \beta \epsilon)V_u(y)=\max_{s}\{U(y,1-s)\epsilon+\lambda s \epsilon \sum_{i=1}^{n}q_i \max [V_u(y),V_e(w;y)]+(1-\lambda s\epsilon)V_u(y)+o(\epsilon)\}$$
Let $s_u^*(y)$ be the optimal searching intensity and solution for unemployment given the level of exogenous non-labor income $y$. Since $\max[a,b]=a+\max[0,b-a]$, we can further rewrite the equation:
$$(1 + \beta \epsilon)V_u(y)=\{U(y,1-s_u^*(y))\epsilon+\lambda s_u^*(y) \epsilon \sum_{i=1}^{n}q_i \max [0,V_e(w;y)-V_u(y)]+V_u(y)+o(\epsilon)\}$$
$$\beta \epsilon V_u(y)=\{U(y,1-s_u^*(y))\epsilon+\lambda s_u^*(y) \epsilon \sum_{i=1}^{n}q_i \max [0,V_e(w;y)-V_u(y)]+o(\epsilon)\}$$
$$\beta V_u(y)=\{U(y,1-s_u^*(y))+\lambda s_u^*(y)\sum_{i=1}^{n}q_i \max [0,V_e(w;y)-V_u(y)]+\lim_{\epsilon \to 0}\frac{o(\epsilon)}{\epsilon}\}$$
$$\implies \beta V_u(y)=U(y,1-s_u^*(y))+\lambda s_u^*(y)\sum_{i=1}^{n}q_i \max [0,V_e(w;y)-V_u(y)]$$

Given knowledge of the function $V_e(w; y)$, this expression in practice is used to simultaneously solve for the optimal search intensity in this state, $s_u^*(y)$, and the value of the unemployment state to an individual with non-labor income $y$, $V_u(y)$.\

For the state of employement with wage $w$ and exogenous non-labor income $y$, we can formulate the Bellman equation in the same manner:
$$V_e(w;y)=\max_{s,l}(1 + \beta \epsilon)^{-1}\{U(y+w(1-s-l),l)\epsilon+\lambda s \epsilon \sum_{i=1}^{n}q_i \max [V_e(\hat{w};y),V_e(w;y)]\\
+\eta \epsilon V_u(y)+\zeta \epsilon V_d(w;y)+(1-\lambda s \epsilon -\eta \epsilon -\zeta \epsilon)V_e(w;y)+o(\epsilon)\}$$

In state of employement, we need to consider job seperation rate $\eta$ and laid-off rate $\zeta$. $\hat{w}$ is the alternative wage offer received given that the individual encounters another potential employer. The choice of whether to accept a new job is extremely simple in this case, since the only characteristic of a job that yields utility is the wage rate. Therefore, the individual will change employers whenever she encounters a higher wage offer. Rearranging terms and passing to the limit, we have
$$(\beta+\eta+\zeta)V_e(w;y)=U(y+w(1-s_e^*(w;y)-l^*(w;y)),l^*(w;y))+\lambda s_e^*(w;y)  \sum_{i=1}^{n}q_i \max [0,V_e(\hat{w};y)-V_e(w;y)]+\eta V_u(y)+\zeta V_d(w;y)$$
This equation includes the optimal choice of leisure consumption and search intensity, $l^*(w;y)$ and $s_e^*(w;y)$ given the wage $w$ and nonlabor income flow $y$. Given the value of $y$, $V_u(y)$ and $V_d(w;y)$, we should simultaneously determines the functions $V_e(w;y)$ and their optimal choices $l^*(w;y)$ and $s_e^*(w;y)$ for each possible wage $w$ in the wage distribution.

Finally, the Bellman equation and value for laid-off state given $w$ and $y$ is
$$V_d(w,y)=\max_{s}(1 + \beta \epsilon)^{-1}\{U(y,1-s)\epsilon+\lambda s \epsilon \sum_{i=1}^{n}q_i \max [V_e(\hat{w};y),V_d(w;y)]+\upsilon \epsilon V_e(w;y)+\eta \epsilon V_u(y)+(1-\lambda s \epsilon-\upsilon \epsilon-\eta \epsilon)V_d(w;y)+o(\epsilon)\}$$
And by taking limit, we have 
$$(\beta+\upsilon+\eta)V_d(w;y)=U(y,1-s_d^*(w;y))+\lambda s_d^*(w;y) \sum_{i=1}^{n}q_i \max [0,V_e(\hat{w};y)-V_d(w;y)]+\upsilon V_e(w;y)+\eta V_u(y)$$

In this equation, $s_d^*(w;y)$ stands for the optimal search intensity when laid off given $(w,y)$, and if we know the value of $V_u(y)$ and $V_e(w;y)$, we should be able to solve the maximization problem and get $s_d^*(w;y)$ and $V_d(w;y)$ simultaneously.

By finishing the last equation, we complete the specification of the BM model. All behavior and equilibrium outcomes are determined
by $V_u(y)$ , $V_e(w;y)$ and $V_d(w;y)$. In our first example, we will set the value of non-labor income $y$ to be fixed, so the equation system could be expressed again without $y$ (as it is given):
$$V_u=U(y,1-s_u^*)+\lambda s_u^*\sum_{i=1}^{n}q_i \max [0,V_e(w)-V_u]\\
(\beta+\eta+\zeta)V_e(w)=U(y+w(1-s_e^*(w)-l^*(w)),l^*(w))+\lambda s_e^*(w)  \sum_{i=1}^{n}q_i \max [0,V_e(\hat{w})-V_e(w)]+\eta V_u+\zeta V_d(w)\\
(\beta+\upsilon+\eta)V_d(w)=U(y,1-s_d^*(w))+\lambda s_d^*(w) \sum_{i=1}^{n}q_i \max [0,V_e(\hat{w})-V_d(w)]+\upsilon V_e(w)+\eta V_u$$

## Numerically Solve BM Model Using Dynamic Programming

We are going to use value iteration and dynamic programming to solve the three Bellman equations simultaneously. But before implementing codes, there are a few things to be specified:

**Utility Function** For the form of $U(c,l)$, we choose to use Cobb-Douglas function with $\alpha =0.6$ and $\beta=0.4$. So the utility function will be $U(c,l)=c^{0.6}l^{0.4}$

**Wage distribution** We utilize the probability density of normal distribution for the wage: there would be higher probability for the wage at the center.

**Wage function** The relationship of working time and income at given wage should be diminishing in margin, and we set a "wage protection" if the individual does not work. So the wage function is $w(1-s-l)=w*(0.1+log_2(1+1-s-l))$. We includ additional 1 inside the log function to prevent negative value, and setting the base to be 2 means that if the individual works full time, she will get $1.1*w$. The use of log function also captures the diminishing return of working hour. Since there is no capital market, the consumption will equal to the income: $c=w(1-s-l)+y=w*(0.1+log_2(1+1-s-l))+y$

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16, 11)  #set default figure size
import numpy as np
from numba import njit, float64
from numba.experimental import jitclass
import quantecon as qe
from quantecon.distributions import BetaBinomial
from scipy.stats import chi2,lognorm
import scipy.stats as sc
```

```python
#n, a, b = 50, 200, 100                        # default parameters for beta-binomial
#q_default = BetaBinomial(n, a, b).pdf()       # default choice of q of betabinomial(won't use here)
#n, df = 50, 10                                #default parameters for chisq: Don't use, will not reach convergence
#x = np.linspace(chi2.ppf(0.01, df),
#                chi2.ppf(0.99, df), n+1)
#q_default = chi2.pdf(x,df)                    #default choice of q of chisq(won't use here)
n,a,b=  50, 35, 5                              #default parameters for normal
w_min, w_max = 10, 60 
w_default = np.linspace(w_min, w_max, n+1)
q_default=sc.norm(a, b).pdf(w_default)         #defualt choice of q
```

```python
fig, ax = plt.subplots()
ax.plot(w_default, q_default, '-o', label='$q(w(i))$')
ax.set_xlabel('wages')
ax.set_ylabel('probabilities')

plt.show()
```

```python
@njit
def u(c, l):    #Here I use Cobb-Douglas with /alpha=0.6 and /beta=0.4
    return ((c**(0.6))*(l**(0.4)))
```

```python
bm_data = [
    ('η', float64),      # job separation rate
    ('β', float64),      # discount factor
    ('y', float64),       # gloabl non-labor income
    ('λ',float64),       # Success rate of searching
    ('ζ',float64),       # Rate of laid-off when employed
    ('υ',float64),       # Rate of recall when laid-off
    ('w', float64[:]),   # list of wage values
    ('q', float64[:])    # pmf of random variable w
]

@jitclass(bm_data)
class BMModel:
    """
    Stores the parameters and functions associated with a given model.
    """

    def __init__(self, λ=0.65, β=0.98, y=5.0,η=0.1, ζ=0.15, υ=0.3, w=w_default, q=q_default):

        self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ = λ, β, y, η, w, q, ζ, υ


    def state_action_values_unemployed(self, Vu, Ve, su):
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ


        Vu_new =(1/β)* (u(y, 1-su) + λ* su * np.sum(np.maximum(0, Ve-Vu) * q))


        return Vu_new
    
    def state_action_values_employed(self, Vu, Ve, Vd, se, L, i):
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ
        
        
        Ve_new=(1/(β+η+ζ))*(u(y+w[i]*(0.1+np.log2(1+1-se-L)), L)+λ* se * np.sum(np.maximum(0, Ve-Ve[i]) * q)+η*Vu+ζ*Vd[i])
        #I use wage function w_received=w*(0.1+log_2(1+h)) while h=1-s-L, the working time. I add 0.1 so that the worker will get  
        #some minimal wages even if he/she did not work.I use log2 to capture the diminishing return of working hour on wage.
         


        return Ve_new
    
    def  state_action_values_laidoff(self, Vu, Ve, Vd, sd, i):
        
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ
        
        
        Vd_new=(1/(β+υ+η))*(u(y, 1-sd)+λ* sd * np.sum(np.maximum(0, Ve-Vd[i]) * q)+η*Vu+υ*Ve[i])


        return Vd_new
    
```

```python
def BellmanVu(bm, Vu, Ve):
        """
        The Bellman equation for Vu
        """
        x_grid=np.linspace(1e-4, 1, 51)
        for i in range(len(x_grid)):
            su = x_grid[i]
            max_val=1e-10
            current_val=bm.state_action_values_unemployed(Vu,Ve,su)
            if current_val > max_val:
                max_val = current_val
        Vu_new=max_val


        return Vu_new
def BellmanVe(bm, Vu, Ve, Vd):
    """
    The Bellman equation for Ve
    """
    x_grid=np.linspace(1e-4, 1, 51)
    Ve_new = np.empty_like(Ve)
    for i in range(len(bm.w)):
        max_val=-1e10
        for se in x_grid:
            for L in x_grid[np.where(x_grid<(1-se))]:
                current_val=bm.state_action_values_employed(Vu, Ve, Vd, se, L, i)
                if current_val > max_val:
                    max_val=current_val
        Ve_new[i]=max_val
    
    return Ve_new


def BellmanVd(bm, Vu, Ve, Vd):
    """
    The Bellman equation for Vd
    """
    x_grid=np.linspace(1e-4, 1, 51)
    Vd_new = np.empty_like(Vd)
    for i in range(len(bm.w)):
        max_val=-1e10
        for sd in x_grid:
            current_val=bm.state_action_values_laidoff(Vu, Ve, Vd, sd, i)
            if current_val > max_val:
                max_val = current_val
        Vd_new[i]=max_val
    
    return Vd_new


def get_greedy(bm, Vu, Ve, Vd):
    """
    Computes the v-greedy policy/optimal searching intensity and leisure of a given function v
    """
    
    x_grid=np.linspace(1e-4, 1, 51)
    se_policy = np.empty_like(x_grid)
    L_policy = np.empty_like(x_grid)
    sd_policy = np.empty_like(x_grid)
    Vu_store = np.empty_like(x_grid)

    for i in range(len(x_grid)):
        max_val_emp=-1e10
        max_val_lof=-1e10

        for se in x_grid:
            for L in x_grid[np.where(x_grid<(1-se))]:
                current_val=bm.state_action_values_employed(Vu, Ve, Vd, se, L, i)
                if current_val > max_val_emp:
                    max_val_emp=current_val
                    max_se = se
                    max_L = L
                    se_policy[i], L_policy[i] = max_se, max_L
                    
        for sd in x_grid:
            current_val_lof=bm.state_action_values_laidoff(Vu, Ve, Vd, sd, i)
            if current_val_lof > max_val_lof:
                max_val_lof = current_val_lof
                max_sd=sd
                sd_policy[i]=max_sd
    
    for j in range(len(x_grid)):
        su=x_grid[j]
        current_val_un=bm.state_action_values_unemployed(Vu, Ve, su)
        Vu_store[j]=current_val_un
    Vu_max=np.max(Vu_store)
    su_policy = x_grid[np.where(Vu_store == Vu_max)]
        
    return se_policy, L_policy, sd_policy, su_policy

```

```python
def solve_model(bm, tol=1e-5, max_iter=2000):
    """
    Iterates to convergence on the Bellman equations

    * bm is an instance of BMModel
    """

    Ve = np.ones_like(bm.w)    # Initial guess of Ve
    Vu = 1                     # Initial guess of Vu
    Vd = np.ones_like(bm.w)    # Initial guess of Vd
    i = 0
    error = tol + 1

    while error > tol and i < max_iter:
        Ve_new=BellmanVe(bm, Vu, Ve, Vd)
        Vu_new=BellmanVu(bm, Vu, Ve)
        Vd_new=BellmanVd(bm, Vu, Ve, Vd)
        error_1 = np.max(np.abs(Ve_new - Ve))
        error_2 = np.abs(Vu_new - Vu)
        error_3 = np.max(np.abs(Vd_new - Vd))
        error = max(error_1, error_2, error_3)
        Ve = Ve_new
        Vu = Vu_new
        Vd = Vd_new
        i += 1
        if i % 25 == 0:
            print(f"Current iteration: {i}")
            print(f"Current distance: {error}")

    return Ve, Vu, Vd
```

Now let us solve the model

```python
bm=BMModel()
Ve, Vu, Vd=solve_model(bm)
se_star, L_star, sd_star, su_star=get_greedy(bm, Vu, Ve, Vd)
import matplotlib.pyplot as plt
Vu_plot=[Vu]*51
su_plot=[su_star]*51
figure, axis = plt.subplots(1, 2)
axis[0].plot(w_default, Vu_plot, label="Umploymed")
axis[0].plot(w_default, Ve, label="Employed")
axis[0].plot(w_default, Vd, label='Laid-off')
axis[0].set_title("Value Function/Utility")
axis[0].set_xlabel("wage")
axis[0].set_ylabel("Utility")
axis[0].legend()
axis[1].plot(w_default, se_star, label="On-job search")
axis[1].plot(w_default, L_star, label='Leisure')
axis[1].plot(w_default, sd_star, label='Laid-off search')
axis[1].plot(w_default, su_plot, label='Unemployment search')
axis[1].set_title("Search Intensity and Leisure")
axis[1].set_xlabel("wage")
axis[1].set_ylabel("Proportion of time")
axis[1].legend()
```

We now have the plots for the value functions, searching intensity and leisure time. There are three things need to be mentioned:
1. The value of employment is strictly higher than unemployment, which means that there is no reservation wage in this default setting.
2. The value and search intensity of unemployment are single values. This is because $Vu$ and $su^*$ are not functions of wages, they are independent of the states of wage.
3. The optimal on-job search intensity $se^*(w)$ is a constant zero, which means that in the default setting, given any wages, the workers will not try to search. This could be the result that the benefits of searching is too few compared with the utility generated by consumption, which is counterintuitive and could suggest some problems. The decreasing laid-off search, leisure time and high unemployment search coincide with our expectation: when wage is high, workers will try to work more to gain more utility and would like to wait for recall if laid-off. Besides, they do not like unemployment and will spend a lot of time searching,

## Model Modification
From the last two parts, we recognized that there were some problems in the utility function setting, causing the on-job searching intensity to be zero at all conditions, and make the leisure decision's partial effects identical across different levels of parameters. After testing the utility function, we found a good alternative and present a modified version for what we done above.

The new utility function is no longer Cobb-Douglas, and it is defined as $u(c,l)=\frac{\sqrt{c}}{10^4}+\frac{\sqrt{10\times l}}{10^4}$. We use square root to capture the diminishing marginal effect, and times 10 before leisure time to leverage the weight of leisure in the utility function. The entire utility function is divided by $10^4$ to make it to be on the same level of searching expectation. 

Besides, we use beta-binomial distribution to simulate a left-skewed distribution, we also adapt linear setting for wage: $w(1-se-l)=w\times(1-se-l)$.

```python
#Modified model
@njit
def uf(c,l):
    return np.sqrt(c)/1e4+np.sqrt(10*l)/1e4 #We first define the utility function. 
                                            #Divide by 1e4 to make it on the same level of searching return.
    
n, a, b = 50, 400, 100                   # default parameters for beta-binomial
q_2 = BetaBinomial(n, a, b).pdf()       # default choice of q of betabinomial
fig, ax = plt.subplots()
ax.plot(w_default, q_2, '-o', label='$q(w(i))$')
ax.set_xlabel('wages')
ax.set_ylabel('probabilities')

plt.show()
```

```python
bmm_data = [
    ('η', float64),      # job separation rate
    ('β', float64),      # discount factor
    ('y', float64),       # gloabl non-labor income
    ('λ',float64),       # Success rate of searching
    ('ζ',float64),       # Rate of laid-off when employed
    ('υ',float64),       # Rate of recall when laid-off
    ('w', float64[:]),   # list of wage values
    ('q', float64[:])    # pmf of random variable w
]

@jitclass(bmm_data)
class BMModel_modified:
    """
    Stores the parameters and functions associated with a given model.
    """

    def __init__(self, λ=0.65, β=0.98, y=5.0,η=0.1, ζ=0.15, υ=0.3, w=w_default, q=q_2):

        self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ = λ, β, y, η, w, q, ζ, υ


    def state_action_values_unemployed(self, Vu, Ve, su):
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ


        Vu_new =(1/β)* (uf(y, 1-su) + λ* su * np.sum(np.maximum(0, Ve-Vu) * q))


        return Vu_new
    
    def state_action_values_employed(self, Vu, Ve, Vd, se, L, i):
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ
        
        
        Ve_new=(1/(β+η+ζ))*(uf(y+w[i]*(1+1-se-L), L)+λ* se * np.sum(np.maximum(0, Ve-Ve[i]) * q)+η*Vu+ζ*Vd[i])
        #I use wage function w_received=w*h while h=1-s-L, the working time.
         


        return Ve_new
    
    def  state_action_values_laidoff(self, Vu, Ve, Vd, sd, i):
        
        #L stands for leisure, su, se stand for searching time when unemployed/employed

        λ, β, y, η, w, q, ζ, υ = self.λ, self.β, self.y, self.η, self.w, self.q, self.ζ, self.υ
        
        
        Vd_new=(1/(β+υ+η))*(uf(y, 1-sd)+λ* sd * np.sum(np.maximum(0, Ve-Vd[i]) * q)+η*Vu+υ*Ve[i])


        return Vd_new
    
```

```python
bm2=BMModel_modified()
Ve_m, Vu_m, Vd_m=solve_model(bm2)
se_star_m, L_star_m, sd_star_m, su_star_m=get_greedy(bm2, Vu_m, Ve_m, Vd_m)
import matplotlib.pyplot as plt
Vu_plot_m=[Vu_m]*51
su_plot_m=[su_star_m]*51
figure, axis = plt.subplots(1, 2)
axis[0].plot(w_default, Vu_plot_m, label="Umploymed")
axis[0].plot(w_default, Ve_m, label="Employed")
axis[0].plot(w_default, Vd_m, label='Laid-off')
axis[0].set_title("Value Function/Utility")
axis[0].set_xlabel("wage")
axis[0].set_ylabel("Utility")
axis[0].legend()
axis[1].plot(w_default, se_star_m, label="On-job search")
axis[1].plot(w_default, L_star_m, label='Leisure')
axis[1].plot(w_default, sd_star_m, label='Laid-off search')
axis[1].plot(w_default, su_plot_m, label='Unemployment search')
axis[1].set_title("Search Intensity and Leisure")
axis[1].set_xlabel("wage")
axis[1].set_ylabel("Proportion of time")
axis[1].legend()
```

This first plot is identical to the default model, and in the second plot, we noticed that on-job search is no longer a constant zero. This means that when the wage is low, the workers will try to search for a better job, and when wage is greater than 18 or 19, they feel it unworthy to find a better job and devote most of time on leisure. The straight line of unemployment search, deacreasing leisure after on-job search intensity hits zero and diminishing laid-off search share the same explainations of the default model.

Now we can conclude what we missed in part (2) in the comparative statics section: higher success rate $\lambda$ will encourage low-income employed people to look for a better place. When the success rate is less than 40%, the individuals will still not search, causing the leisure decision to be the same(just like in previous section), but with higher success rate, the individual with low wages will search more and spend less time on leisure in order to search. For high-income people, they do not search at all so they converge to the same leisure dicision.

The interpretation for job separation rate $\eta$'s effect is opposite to the success rate $\lambda$. Workers will only search when both income and separation rate are low, and lower $\eta$ will bring less leisure time and higher search intensity. The possible reason is that, when job separation rate is low, individuals have fewer chances of being fired, and low-income workers would like to find a better job in a stable market and hence spend more time on search and less time on leisure. But when job separation rate is high (greater than 60% perhaps), the workers are not willing to search anymore as the great volatility depreciates the value of finding a good job, which will cause them to transfer the searching time to leisure. The graphs provide satisfactory results and fit our expectation.

## Possible Improvements
There are some directions that could follow to further improve the performance of the model:
1. Though our model works well on normal distribution and beta-binomial distribution, it does not incorporate well with right-skewed distribution, such as exponential or chi-squar distribution,while in the orginal paper, Burdett and Mortensen used exponential to study the effect of hazard rate. So one way to improve the model is to make it a better fit for a right-skewed wage distribution.
2. The calculation of the dynamic model is immense. Though we used JIT algorithm for the BM model instance, we fail to apply JIT to the functions that calculate the Bellman equations and the value iteration function. Though introducing the JIT Iinto the BM model instance has greatly speeded up the calculation, there are still a lot of space for further efficiency improvement, and this is the second way to enahnce the performance our model.
